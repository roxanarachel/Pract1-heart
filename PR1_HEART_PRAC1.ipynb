### ***Contexto*** ###

Hoy en la actualidad las enfermedades cardiovasculares (ECV) son la principal causa de muerte a nivel mundial y se estima que cobran 17,9 millones de vidas cada año, lo que representa el 31% de todas las muertes en todo el mundo. Cuatro de cada cinco muertes por ECV se deben a ataques cardíacos y accidentes cerebrovasculares, y un tercio de estas muertes ocurren prematuramente en personas menores de 70 años. Los principales síntomas que causa la insuficiencia cardiaca es la disminución del aporte de oxígeno a los tejidos, debido a un descenso del gasto cardiaco y los mecanismos compensadores, que tienden aumentar las presiones intravasculares y provocar síntomas de congestión.

La insuficiencia cardíaca es un evento común causado por enfermedades cardiovasculares, es por eso que este conjunto de datos contiene 11 características que pueden usarse para predecir una posible enfermedad cardíaca.

##### ***Acerca del conjunto de datos*** #####

***_Age_***: Edad del paciente [años].
***_Sex_***: Sexo del paciente [M: Masculino, F: Femenino].
***_ChestPainType_***: Tipo de dolor en el pecho [TA: angina típica, ATA: angina atípica, NAP: dolor no anginoso, ASY: asintomático].
***_RestingBP_***: Presión arterial en reposo [mm Hg].
***_Colesterol_***: Colesterol sérico [mm/dl].
***_FastingBS_***: Azúcar en sangre en ayunas [1: si FastingBS > 120 mg/dl, 0: en caso contrario].
***_RestingECG_***: Resultados del electrocardiograma en reposo [Normal: normal, ST: con anomalía de la onda ST-T (inversiones de la onda T y/o elevación o depresión del ST > 0,05 mV).
HVI: muestra hipertrofia ventricular izquierda probable o definitiva según los criterios de Estes].
***_MaxHR_***: Frecuencia cardíaca máxima alcanzada [Valor numérico entre 60 y 202].
***_ExerciseAngina_***: Angina inducida por el ejercicio [Y: Sí, N: No].
***_Oldpeak_***: oldpeak = ST [Valor numérico medido en depresión].
***_ST_Slope_***: La pendiente del segmento ST del ejercicio máximo [Up: ascendente, Flat: plano, Down: descendente].
***_HeartDisease_***: clase de salida [1: enfermedad cardíaca, 0: normal].

Esta línea es utilizada para la biblioteca Seaborn, nos genera una gráfica de barras a partir de la frecuencia de los datos en ***RestingECG***, designando cada uno de los grupos de diagnóstico que son ***Normal***, ***ST*** y ***LVH*** indicados por los resultados del electrocardiograma en reposo.

Por otro lado en este código se utiliza Matplotlib para la creación de un gráfico de dispersión con los datos ***MaxHR*** y ***Age***, como se puede visualizar los datos bidimensionales se focalizan en las coordenadas del eje=X con datos de frecuencia cardíaca máxima alcanzada de 80 a 170 y en el eje=Y la edad de los individuos 50 a 70 años en un espacio de dispersión.

```python
x = df['MaxHR']  
y = df['Age']   

sizes = np.random.uniform(5, 40, len(df['MaxHR']))
colors = np.random.uniform(10, 100, len(df['MaxHR']))

fig, ax = plt.subplots()

ax.scatter(x, y, s=sizes, c=colors, vmin=0, vmax=100)


plt.xlabel('MaxHR')  
plt.ylabel('Age')   

ax.set(xlim=(50, 200), ylim=(0, 90)) 
plt.show()
```

En esta línea de código se utiliza Pandas para generar una tabla de datos reales del archivo CSV de su totalidad 918 datos para mapear los valores de ***HeartDisease*** de diagnóstico final, siendo separados por ***Sex*** (género) y juntos para el diagnóstico de cada grupo de individuos ya sea ***Enfermedad*** o **Normal***.

```python
data = {
    "HeartDisease": [1, 0, 1, 0, 1, 0, 1, 0, 1] * 102,  
    "Sex": ["M", "F", "M", "F", "M", "F", "M", "M", "F"] * 102
}

df = pd.DataFrame(data)

df["HeartDisease"] = df["HeartDisease"].map({1: "Enfermedad", 0: "Normal"})

tabla_contingencia = pd.crosstab(index=df["Sex"], columns=df["HeartDisease"], margins=True, margins_name="Total")

tabla_contingencia
```
